```python
# imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import pathlib
import yaml

# PCA
from sklearn.decomposition import PCA


%matplotlib inline
```

```python
# load data
CONFIG_FILE = '../experiments/train_mnist.yaml'
RUN_NAME = 'mnist_mlp_20230607-060024'

# 'mnist_mlp_20230607-054432'

# load config
with open(CONFIG_FILE) as f:
    config = yaml.safe_load(f)

# get data
save_dir = config["save_dir"]

# save_dir actually starts with ../
save_dir = '../' + save_dir

model_params = config["model"]
dataset_params = config["data"]
training_params = config["training"]
logging_params = config["logging"]

# get saving directories
save_dir = pathlib.Path(save_dir)

run_dir = save_dir / RUN_NAME

model_save_dir = run_dir / model_params["model_save_dir"]
model_save_path = model_save_dir / (RUN_NAME + ".ckpt")

activations_save_dir = run_dir / model_params["activations_save_dir"]
activations_save_path = activations_save_dir / ("activations" + ".h5")

gradients_save_dir = run_dir / model_params["gradients_save_dir"]
gradients_save_path = gradients_save_dir / ("gradients" + ".h5")

weight_gradients_save_dir = run_dir / model_params["weights_save_dir"]
weight_gradients_save_path = weight_gradients_save_dir / ("weight_gradients" + ".h5")

weights_save_dir = run_dir / model_params["weights_save_dir"]
weights_save_path = weights_save_dir / ("weights" + ".h5")

logger_csv = model_save_dir / 'lightning_logs' / 'version_0' / 'metrics.csv'
```

```python
# load data
df = pd.read_csv(logger_csv)

# select rows where train_loss_step is not nan
train_step_df = df[df['train_loss_step'].notna()]
valid_step_df = df[df['val_loss_step'].notna()]

# select rows where train_loss_epoch is not nan
train_epoch_df = df[df['train_loss_epoch'].notna()]
valid_epoch_df = df[df['val_loss_epoch'].notna()]

max_step = max(train_step_df['step']) + 1
max_epoch = int(max(train_epoch_df['epoch'])) + 1
max_step, max_epoch
```

### Plotting weight space

```python
# load the weight matrices as an .h5 file
import h5py

split = 'train'
layer_names = ['fc_layers.{i}.weight'.format(i=i) for i in range(3)]

weight_dict = {layer_name: [] for layer_name in layer_names}

with h5py.File(weights_save_path, 'r') as f:
    for epoch_num in range(max_epoch):
        epoch_group = f'epoch_{epoch_num:03d}'
        # print(epoch_group)
        for layer_name in layer_names:
            # get the weights for this layer
            weights = np.array(f[epoch_group][split][layer_name])

            # append to the list
            weight_dict[layer_name].append(weights)

# convert to numpy arrays
for layer_name in layer_names:
    weight_dict[layer_name] = np.array(weight_dict[layer_name])
    # reshape to (epoch, row * col)
    weight_dict[layer_name] = weight_dict[layer_name].reshape(max_epoch, -1)
    print(weight_dict[layer_name].shape)
```

```python
# elbow method to determine best k for PCA
from sklearn.decomposition import PCA

elbow_dict = {layer_name: [] for layer_name in layer_names}
elbow_dict['k'] = [1, 2, 3, 4, 5, 10, 20, 50]

for layer_name in layer_names:
    for k in elbow_dict['k']:
        pca = PCA(n_components=k)
        pca.fit(weight_dict[layer_name])
        print(k, pca.explained_variance_ratio_.sum())

        elbow_dict[layer_name].append(pca.explained_variance_ratio_.sum())
```

```python
# plot elbow method
fig, ax = plt.subplots(1, 3, figsize=(15, 5))

for i, layer_name in enumerate(layer_names):
    ax[i].plot(elbow_dict['k'], elbow_dict[layer_name])
    ax[i].set_title(layer_name)
    ax[i].set_xlabel('k')
    ax[i].set_ylabel('explained variance ratio')
    ax[i].set_ylim(0, 1)
    ax[i].grid()
```

```python
# fitting PCA
k = 2

# separate PCA for each layer
pca_dict = {layer_name: PCA(n_components=k) for layer_name in layer_names}

# fit PCA
for layer_name in layer_names:
    pca_dict[layer_name].fit(weight_dict[layer_name])

```

```python
# plot PCA
fig, ax = plt.subplots(1, 3, figsize=(15, 5))

for i, layer_name in enumerate(layer_names):
    ax[i].scatter(
        pca_dict[layer_name].components_[0],
        pca_dict[layer_name].components_[1],
        alpha=0.5,
    )
    ax[i].set_title(layer_name)
    ax[i].set_xlabel('PC1')
    ax[i].set_ylabel('PC2')
    ax[i].grid()
```

```python
# plot trajectory of PCA over epochs, as a video, with a red dot for the current epoch. past epochs are in grey, future epochs are in blue.
from matplotlib.animation import FuncAnimation
from IPython.display import HTML

fig, ax = plt.subplots(1, 3, figsize=(15, 5))

# plot the PCA
for i, layer_name in enumerate(layer_names):
    ax[i].scatter(
        pca_dict[layer_name].components_[0],
        pca_dict[layer_name].components_[1],
        alpha=0.5,
        color='grey',
    )
    ax[i].set_title(layer_name)
    ax[i].set_xlabel('PC1')
    ax[i].set_ylabel('PC2')
    ax[i].grid()

# plot the current epoch in red
epoch = 0
epoch_lines = []
for i, layer_name in enumerate(layer_names):
    epoch_lines.append(
        ax[i].plot(
            pca_dict[layer_name].components_[0][epoch],
            pca_dict[layer_name].components_[1][epoch],
            marker='o',
            color='red',
        )[0]
    )

# plot the future epochs in blue
for i, layer_name in enumerate(layer_names):
    epoch_lines.append(
        ax[i].plot(
            pca_dict[layer_name].components_[0][epoch+1:],
            pca_dict[layer_name].components_[1][epoch+1:],
            marker='o',
            color='blue',
        )[0]
    )

# plot the past epochs in grey
for i, layer_name in enumerate(layer_names):
    epoch_lines.append(
        ax[i].plot(
            pca_dict[layer_name].components_[0][:epoch],
            pca_dict[layer_name].components_[1][:epoch],
            marker='o',
            color='grey',
        )[0]
    )

# function to update the plot
def update(epoch):
    for i, layer_name in enumerate(layer_names):
        epoch_lines[i].set_data(
            pca_dict[layer_name].components_[0][epoch],
            pca_dict[layer_name].components_[1][epoch],
        )
    return epoch_lines

# animate
anim = FuncAnimation(fig, update, frames=np.arange(max_epoch), interval=100)

# save to gif
anim.save('pca.gif', writer='imagemagick')
```

```python
import io
from PIL import Image

def fig2img(fig):
    """
    Convert a Matplotlib figure to a PIL Image and return it
    https://stackoverflow.com/a/61754995
    """
    buf = io.BytesIO()
    _ = fig.savefig(buf)
    _ = buf.seek(0)
    img = Image.open(buf)
    img = img.convert('RGB')
    return img
```

```python
# PSD on the axes of variance

import os
import scipy
# get the power spectral density of the pca of the activations
os.makedirs('psd', exist_ok=True)

k = 10
psd_trajectory_dict = {}
for name, activations in weight_dict.items():
    # get the pca of the activations to k dimensions
    pca = PCA(n_components=k)
    pca.fit(activations)
    activations_pca = pca.transform(activations)  # (num_episodes_to_run * num_steps_per_episode, k)

    # activations_reshaped = activations_pca.reshape(num_episodes_to_run, num_steps_per_episode,
    #                                                -1)  # (episodes, steps, k)
    f, psd = scipy.signal.welch(activations_pca, fs=1, nperseg=50, noverlap=25, axis=-2)

    psd_pca = np.average(psd, axis=-1, weights=pca.explained_variance_ratio_)  # (episodes, steps)
    psd_trajectory = psd_pca
    # psd_trajectory = np.mean(psd_pca, axis=0)  # (steps,)

    psd_trajectory_dict[name] = psd_trajectory

    # make a plot of the power spectral density
    fig, ax = plt.subplots()
    ax.semilogy(f, psd_trajectory)
    ax.set_title(name)
    ax.set_xlabel('frequency [Hz]')
    ax.set_ylabel('PSD [V**2/Hz]')
    # set x scale to log
    ax.set_xscale('log')

    save_path = 'psd/{}.png'.format(name)
    # save_path = None
    fig_im = fig2img(fig)
    if save_path is not None:
        fig_im.save(save_path)

# make a plot of the power spectral density of all the layers from psd_trajectory_dict
fig, ax = plt.subplots()
for name, psd_trajectory in psd_trajectory_dict.items():
    ax.semilogy(f, psd_trajectory, label=name)
ax.set_title('All layers')
ax.set_xlabel('frequency [Hz]')
ax.set_ylabel('PSD [V**2/Hz]')
# set x scale to log
ax.set_xscale('log')
ax.legend()

save_path = 'psd/all_layers.png'
# save_path = None
fig_im = fig2img(fig)
if save_path is not None:
    fig_im.save(save_path)
```
